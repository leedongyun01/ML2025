```python
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")
sns.set_style("whitegrid")
sns.lineplot(x = "timepoint", y = "signal", data = fmri)
```

```python
fmri.sample(n = 10, random_state = 1)
```

```python
sns.lineplot(x = "timepoint", y = "signal", hue = "event", data = fmri)
```

```python
# 회귀 그래프 regplot
tips = sns.load_dataset("tips")
sns.regplot(x = "total_bill", y = "tip", data = tips, x_ci = 95)
```

```python
# 산점도 scatterplot
tips = sns.load_dataset("tips")
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
```

```python
# 비교 그래프 countplot
tips = sns.load_dataset("tips")
sns.countplot(x="smoker", hue="time", data=tips)
```

```python
# 막대그래프 barflot
sns.barplot(x="day", y="total_bill", data=tips)
```

```python
# 바이올린 그래프 violinplot
sns.violinplot(x="day", y="total_bill",
hue="smoker",data=tips, palette="muted")
```

```python
# 스웜 플롯 swarmplot
sns.swarmplot(x="day", y="total_bill", hue="smoker",
data=tips, palette="muted")
```

```python
# 패싯 그리드 FacetGrid
g = sns.FacetGrid(tips, col="time",
row="sex")
```

```python
g = sns.FacetGrid(tips,
col="time",
row="sex")
g.map(sns.scatterplot,
"total_bill", "tip")
```

```python
g = sns.FacetGrid(tips, col="time",
row="sex")
g.map_dataframe(sns.histplot,
x="total_bill")
```
