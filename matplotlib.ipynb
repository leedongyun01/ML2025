```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.arange(100)
Y = np.arange(100)

plt.plot(X, Y)
plt.show()
```

```python
X_1 = np.arange(30)
Y_1 = [np.cos(value) for value in X_1]
       
X_2 = np.arange(30)
Y_2 = [np.sin(value) for value in X_1]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
fig, ax = plt.subplots()

X_1 = np.arange(100)
Y_1 = [np.cos(value) for value in X]

ax.plot(X_1, Y_1)
ax.set(title = 'cos graph',
       xlabel = 'X',
       ylabel = 'Y')

plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1, c = "b")
ax_2.plot(X_2, Y_2, c = "g")
plt.show()
```

```python
x = np.linspace(-1, 1, 100)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(x, y_1)
ax[0, 1].plot(x, y_2)
ax[1, 0].plot(x, y_3)
ax[1, 1].plot(x, y_4)

plt.show()
```

```python
X_1 = np.arange(100)
Y_1 = [value for value in X]
X_2 = np.arange(100)
Y_2 = [value + 100 for value in X]

plt.plot(X_1, Y_1, color="#000000") 
plt.plot(X_2, Y_2, c="c")

plt.show()
```

```python
plt.plot(X_1, Y_1, c = "b", linestyle = "dashed")
plt.plot(X_2, Y_2, c = "r", ls = "dotted")

plt.title("Two Lines")
plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1, c = "b")
ax_1.set_title("Figure 1")
ax_2.plot(X_2, Y_2, c = "g")
ax_2.set_title("Figure 2")

plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=False, loc="upper right")
# 범례 legend

plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 

plt.scatter(x, y, s=area, c=colors, alpha=0.5) # 산점도 scatter
plt.show()
```

```python
data = [[5., 25., 50., 20.],
[4., 23., 51., 17],
[6., 22., 52., 19]]

X = np.arange(0,8,2)

# 막대그래프 bar
plt.bar(X + 0.00, data[0], color = 'r', width = 0.50)
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(X + 1.0, data[2], color = 'b', width = 0.50)

# x축의 위치, 이름
plt.xticks(X+0.50, ("A","B","C", "D"))

plt.show()
```

```python
 data = np.array([[5., 25., 50., 20.],
[4., 23., 51., 17],
[6., 22., 52., 19]])

color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])

# 누적 막대그래프
for i in range(3):
    plt.bar(X, data[i], 
        bottom = np.sum(data[:i], axis=0),
        color = color_list[i], 
        label = data_label[i])
    
plt.legend()
plt.show()
```
